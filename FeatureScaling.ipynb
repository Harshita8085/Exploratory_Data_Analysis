```python
# 1.Standarization-Is is very effective technique which re-scale a feture value so that
# it has distibution with  0 mean value and variance equals to 1.
# Xnew=(Xi-Xmean)/standard_deviation
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df = pd.read_csv("loan_data_set.csv")
df.head(3)
```

```python
df.describe()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df["CoapplicantIncome"].fillna(df["CoapplicantIncome"].mean(),inplace=True)
```

```python
sns.distplot(df["CoapplicantIncome"])
plt.show()
```

```python
df.describe()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
```

```python
ss.fit(df[["CoapplicantIncome"]])
```

```python
df["CoapplicantIncome_ss"]=pd.DataFrame(ss.transform(df[["CoapplicantIncome"]]))
```

```python
df.head(5)
```

```python
df.describe()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
sns.distplot(df["CoapplicantIncome"])
plt.title("Before")

plt.subplot(1,2,2)
sns.distplot(df["CoapplicantIncome_ss"])
plt.title("After")

plt.show()
```

```python
# 2.Normalization- It is scaling technique in which values are shifted and rescaled 
# so that they end up ranging between 0 and 1.It is also called Min Max Scaling
# Xnew=(Xi-Min(x))/Max(x)-Min(x)
```

```python
data=pd.read_csv("loan_data_set.csv")
data.head(3)
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mn=MinMaxScaler()
```

```python
mn.fit(data[["ApplicantIncome"]])
```

```python
data["ApplicantIncome_mn"]=pd.DataFrame(mn.transform(data[["ApplicantIncome"]]))
```

```python
data.head(3)
```

```python
data.describe()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title("Before")
sns.distplot(data["ApplicantIncome"])

plt.subplot(1,2,2)
plt.title("After")
sns.distplot(data["ApplicantIncome_mn"])

plt.show()
```
