```python
# two type of categorical data- 1.nominal , 2.ordinal
# nominal- not in any sequence or order
# ordinal- present in order 
```

```python
# encoding of Nominal data
from sklearn.preprocessing import LabelEncoder
import pandas as pd
```

```python
df = pd.DataFrame({"Name":["Dog","Cat","Cow","Black","wscube"]})
df
```

```python
le=LabelEncoder()
```

```python
df["en_Name"]=le.fit_transform(df["Name"])
```

```python
df
```

```python
# ordinal encoding - 1.sklearn 2.map function
# 1)ordinal encoding using sklearn librabry
data=pd.DataFrame(
    {
        "sizes" : ["s","m","l","xl","s","m","l","s","s","l","xl","m"]
    }
)
data.head(5)
```

```python
# ord_data=[["s","m","l","xl"]]
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
encod=OrdinalEncoder(categories=[["s","m","l","xl"]])
```

```python
data.info()
```

```python
data["encod_size"]=encod.fit_transform(data[["sizes"]])
```

```python
data
```

```python
# 2)ordinal encoding using map function
ord_data1 = {"s":0,"m":1,"l":2,"xl":3}
data["end_ordinal"]=data["sizes"].map(ord_data1)
```

```python
data
```
